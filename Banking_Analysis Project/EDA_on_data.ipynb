```python
#1) want to convert (ESTIMATED INCOME) to categorical column
bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels,right=False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique categories in categorical columns
categorical_cols = df[["Amount of Credit Cards","BRId","GenderId","Properties Owned",
                       "Occupation","Fee Structure","Risk Weighting","Nationality",
                       "IAId","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for {col}:")
    display(df[col].value_counts())
```

```python
#visualize this using charts   --- UNIVARIATE ANALYSIS-- one column at a time
for i ,predictor in enumerate(df[["Amount of Credit Cards","BRId","GenderId","Properties Owned",
                       "Occupation","Fee Structure","Risk Weighting","Nationality",
                       "IAId","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)                              
```

## BIVARIATE ANALYSIS -- analyze one column wrt to another column

```python
for i ,predictor in enumerate(df[["Amount of Credit Cards","BRId","GenderId","Properties Owned",
                       "Occupation","Fee Structure","Risk Weighting","Nationality",
                       "IAId","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue="GenderId") 
```

```python
# UNIVARIATE ANALYSIS and visualization
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance',
                 'Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts'
                 ,'Foreign Currency Account','Business Lending']
plt.figure(figsize=(16,30))
for i,col in enumerate(numerical_cols):
    plt.subplot(5,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()    
```

## HEATMAPS -- Correlation tells you how two variables are related.

```python
numerical_cols = [
    'Estimated Income','Superannuation Savings','Credit Card Balance',
    'Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts',
    'Foreign Currency Account','Business Lending'
]
correlation_matrix = df[numerical_cols].corr()  #correlation is done
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## INSIGHTS FROM EDA

1) the strongest positive correlation occurs between "bank deposits" with "checking accounts"
,"saving accounts"  and "foreign currency accounts" indicating that customers who maintain high
balances in one account type often hold substanial amount/funds in other accounts as well.
